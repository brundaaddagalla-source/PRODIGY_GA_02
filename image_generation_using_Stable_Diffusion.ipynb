```python
import torch
from diffusers import StableDiffusionPipeline
import gradio as g
```

```python
model="sd-legacy/stable-diffusion-v1-5"
pipe=StableDiffusionPipeline.from_pretrained(
    model, torch_dtype=torch.float16
)
pipe=pipe.to("cuda")
```

```python
def generate(p):
  image=pipe(p).images[0]
  return image
```

```python
i=g.Interface(
    fn=generate,
    inputs=g.Textbox(
        placeholder="What image would you like to Generate?",
        label="text_prompt"
    ),
    outputs=g.Image(label="Generated image"),
    title="Prodigy InfoTech- Text to Image Generator using Stable Diffusion",
)
i.launch()
```
